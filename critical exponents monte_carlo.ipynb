<h2>Pendiente de ubinder </h2>
Pendiente calculada en el valor de temperatura critica $T_c = 2.28$

```python
from pylab import *
T_C = "2.28"
tc = float(T_C)

def create_func(x,d):
    k = {str(x[i]): d[i] for i in range(len(d))}
    return k

def derive(x,y,order = "1"):
    if order == "1":
        dx = x[1] - x[0]
        return (y[1:] - y[:-1])/dx
    elif order == "2":
        dx = 2*(x[1] - x[0])
        return (y[2:]- y[:-2])/dx

def load_data(fname,cols = (0,1)):
    data = loadtxt(fname,usecols= cols)
    
    X = []
    for i in range(len(cols)):
        X.append(data[:,i])
    return X
```

Entonces las respectivas pendientes de Ubinder para $L = \{ 8,10,12,14,18,24,32\}$ son:

```python
ms = []
for file_name in files1:
    x = load_data(dirname + file_name,cols=(0,5))
    f = create_func(x[0],derive(x[0],x[1]))
    ms.append(-1*f[T_C])
ms
```

de la expresion 
$$\displaystyle -\frac{dU_B}{dT} \approx L^{1/\nu}$$
Se determina el exponente crítico  $$\nu = 1.05390902286$$

```python
from scipy import stats

xlog = log(sizes)
ylog  = log(ms)

m,b,r,p,std = stats.linregress(xlog,ylog)
nu = 1./m

x = sizes
y = ms
z = exp(m*xlog+b)

data1 = [tuple([x[i],y[i]]) for i in range(len(x))]
data2 = [tuple([x[i],z[i]]) for i in range(len(x))]

g3 = graph.graphxy(width=8,height=5.5,
                   x=graph.axis.log(min = 5e0,max = 0.5e2,title=r"$L$"),
                   y=graph.axis.log(min = 1e-1,max = 1e0,title=r"$-dU/dT$"),
                   key = graph.key.key(pos = "tl"))

g3.plot(graph.data.points(data1,x = 1,y = 2,title = "data"),
        [graph.style.symbol(symbol =  graph.style.symbol.changecircletwice,size = 0.17)])

g3.plot(graph.data.points(data2,x = 1,y = 2,title = r"$\nu = %s \pm %s$"%(str(nu)[:5],str(std)[:5])),
        [graph.style.line([style.linestyle.dashdotted])])
g3.writePDFfile("figures/nu")

print "nu = ",nu,"+-",std
g3
```

Magnetización promedio por sitio $\langle M \rangle$ a  $T_c$  
$$\beta =  0.135784417106$$

```python
M_prom = []
for file_name in files1:
    L = float(file_name[3:5])
    x = load_data(dirname + file_name,cols=(0,2))
    f = create_func(x[0],x[1])
    M_prom.append(f[T_C]*(1/L**2))
    
xlog = log(sizes)
ylog  = log(M_prom)

m,b,r,p,std = stats.linregress(xlog,ylog)

beta = -nu*m
print "beta = ",beta,"+-",std

x = sizes
y = M_prom
z = exp(m*xlog+b)

data1 = [[x[i],y[i]] for i in range(len(x))]
data2 = [[x[i],z[i]] for i in range(len(x))]

g4 = graph.graphxy(width=8,height=5.5,
                   x=graph.axis.log(min = 7e0,max = 0.35e2,title=r"$L$"),
                   y=graph.axis.log(min = 0.6e0,max = 0.9e0,title=r"$\langle M_T \rangle$"),
                   key = graph.key.key(pos = "tr"))

g4.plot(graph.data.points(data1,x = 1,y = 2,title = "data"),
        [graph.style.symbol(symbol =  graph.style.symbol.circle,size = 0.17)])

g4.plot(graph.data.points(data2,x = 1,y = 2,title = r"$\beta = %s \pm %s$"%(str(beta)[:5],str(std)[:5])),
        [graph.style.line([style.linestyle.dashdotted])])
g4.writePDFfile("figures/beta")
g4
```


Reescalamiento de las curvas de magnetización

```python
g = graph.graphxy(width=8,height=5.5,
                  x = graph.axis.linear(min = -3,max = 3,title = r"$L^{1/\nu}(T-T_c)/T_c$"),
                  y = graph.axis.linear(min = 0.2,max = 1.7,title = r"$\langle M_T \rangle L^{\beta/\nu}$"),
                  key = graph.key.key(pos = "tr"))

for j in range(len(sizes)):
    l = sizes[j]
    x = load_data(dirname + files2[j] ,cols = (0,2))
    T = [l**(1./nu)*(t-tc)/tc for t in x[0]]
    M = [l**(beta/nu)*m*(1.0/(l*l)) for m in x[1]]
    dat = [tuple([T[i],M[i]]) for i in range(len(T))]
    
    g.plot(graph.data.points(dat, x=1, y=2, title = r"\scriptsize $L = {}$".format(sizes[j])),
           [graph.style.line([colours[j],style.linestyle.solid,style.linewidth.thick])])

g.writePDFfile("figures/magnetiza_rescale")
g
```

Suceptibilidad magnética $\chi_T$, magnetización promedio cuadrado $\langle M_T\rangle^2$, y suceptibilidad máxima a $T_c$ por sitio
$$\gamma = 1.87479734338$$

```python
def regression(x,y):
    xlog = log(x); ylog = log(y)
    m,b,r,p,std = stats.linregress(xlog,ylog)
    return m,exp(m*xlog+b),std

Ms = []; Xs = []; Xs_max = []
for j in range(len(sizes)):
    l = sizes[j]
    x = load_data(dirname + files1[j] ,cols = (0,4,6))
    T = array(x[0]); X = array(x[1]); M= array(x[2])
    f1 = create_func(T,X); f2 = create_func(T,M);
    
    Ms.append(f2[T_C]*(1.0/(l*l)))
    Xs.append(f1[T_C]*(1.0/(l*l)))
    Xs_max.append(X.max()*(1.0/(l*l)))

m1,z1,std1 = regression(sizes,Ms)
m2,z2,std2 = regression(sizes,Xs)
m3,z3,std3 = regression(sizes,Xs_max)

data1 = [[sizes[i],Ms[i]] for i in range(len(sizes))]
data2 = [[sizes[i],Xs[i]] for i in range(len(sizes))]
data3 = [[sizes[i],Xs_max[i]] for i in range(len(sizes))]
lin1 = [[sizes[i],z1[i]] for i in range(len(sizes))]
lin2 = [[sizes[i],z2[i]] for i in range(len(sizes))]
lin3 = [[sizes[i],z3[i]] for i in range(len(sizes))]

g6 = graph.graphxy(width = 8,height=5.5,
                   x=graph.axis.log(min = 0.6e1,max = 0.4e2,title=r"$L$"),
                   y=graph.axis.log(min = 0.5e0,max = 4.5e3,title=r""),
                   key = graph.key.key(pos = "tl"))

g6.plot(graph.data.points(data1,x = 1,y =2,title = r"$\gamma/\nu = %s \pm %s$"%(str(m1)[:5],str(std1)[:5])))

g6.plot(graph.data.points(data2,x = 1,y =2,title = r"$\gamma/\nu = %s \pm %s$"%(str(m2)[:5],str(std2)[:5])))

g6.plot(graph.data.points(data3,x = 1,y =2,title = r"$\gamma/\nu = %s \pm %s$"%(str(m3)[:5],str(std3)[:5])))

g6.plot(graph.data.points(lin1,x = 1,y =2,title = None),
        [graph.style.line([style.linestyle.dashdotted])])

g6.plot(graph.data.points(lin2,x = 1,y =2,title = None),
        [graph.style.line([style.linestyle.dashdotted])])

g6.plot(graph.data.points(lin3,x = 1,y =2,title = None),
        [graph.style.line([style.linestyle.dashdotted])])

g6.writePDFfile("figures/gamma")

gamma1 = m1*nu
gamma2 = m2*nu
gamma3 = m3*nu
gamma = (gamma1 + gamma2 + gamma3)/3.
print "gamma = ",gamma
g6
```

Reescalamiento de las curvas de suceptibilidad

```python
g = graph.graphxy(width=8,height=5.5,
                  x = graph.axis.linear(max = 10,title = r"$L^{1/\nu}(T-T_c)/T_c$"),
                  y = graph.axis.linear(title = r"$\chi L^{-\gamma/\nu}$"),
                  key = graph.key.key(pos = "tr"))

for j in range(len(sizes)):
    l = sizes[j]
    x = load_data(dirname + files2[j] ,cols = (0,4))
    T = [l**(1./nu)*(t-tc)/tc for t in x[0]]
    X = [l**(-gamma/nu)*xs*(1.0/(l*l)) for xs in x[1]]
    dat = [[T[i],X[i]] for i in range(len(T))]
    
    g.plot(graph.data.points(dat, x=1, y=2, title = r"\scriptsize $L = {}$".format(sizes[j])),
           [graph.style.line([colours[j],style.linestyle.solid,style.linewidth.thick])])

g.writePDFfile("figures/suceptibility_rescale")
g
```
